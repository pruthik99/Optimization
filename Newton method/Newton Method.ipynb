```python
#(a)
# for first condition
# |f(xk)| < Ea

def newton(f,D_f,x0,epsilon,iteration):
    
    xn = x0
    for n in range(0,iteration):
        fxk = f(xn)
        if abs(fxk) < epsilon:
            print('Found solution after',n,'iterations.\n')
            return xn
        D_fxk = D_f(xn)
        if D_fxk == 0:
            print('No root for zero derivative.')
            return None
        xn = xn - fxk/D_fxk
    print('Exceeded maximum number of iterations. No solution found.')
    return None

e = 2.71
Fun = lambda x: e**-x - 10**-9
Der_Of_Fun = lambda x: -e**-x
root = newton(Fun,Der_Of_Fun,0,0.001,100)
print('The root of given f(x) is:',root)
```

```python
#(b)

# first condition ||Xk+1 - Xk|| < 0.0001

from sympy import symbols
import numpy as np

x,y = symbols('x y') #here we take x1=x and x2=y

fun = (x-1)**4 + (2*(x-1)**2)*((y-1)**2) + ((y+1)**4) - (2*(y-1)**2) - (2*y+1)**2 +1
x0 = np.array([[1.21,-1.15]])
e = 0.001
iteration = 100

df_x = fun.diff(x)
df_y = fun.diff(y)
df_xx = fun.diff(x,x)
df_yy = fun.diff(y,y)
df_xy = fun.diff(x,y)   
df_yx = fun.diff(y,x)

def newton(x0,e,iteration):
    counter = 1
    flag = 1
    stopping_criteria = True
    
    while stopping_criteria:
        gradient = np.array([df_x.subs({x:x0[0][0],y:x0[0][1]}).evalf(),
                     df_y.subs({x:x0[0][0],y:x0[0][1]}).evalf()])
        
        hessian = np.matrix([[df_xx.subs({x:x0[0][0],y:x0[0][1]}).evalf(),df_yx.subs({x:x0[0][0],y:x0[0][1]}).evalf()],
                        [df_xy.subs({x:x0[0][0],y:x0[0][1]}).evalf(),df_yy.subs({x:x0[0][0],y:x0[0][1]}).evalf()]],dtype=float)
        
        x1 = np.array(x0 - np.array(np.dot(gradient,np.linalg.inv(hessian))),dtype=float)
        print('Iteration: ',counter)
        counter = counter +1
        
        if counter > iteration:
            flag = 0
            break
            
        print(x0,x1)
        stopping_criteria = np.linalg.norm(x1-x0) > e
        x0=x1
        
    if flag == 1:
        print('\n\n\nExtremum point for given f(x) is: ',x0)
    else:
        print('\nNot found any extremum point within given iteration.')
        
newton(x0,e,iteration)
```
